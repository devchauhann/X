# AIMMR Final - Quick Test Notebook
This notebook tests the core modules in aimmr_core.

```python
from aimmr_core import PatternAnalyzer, CognitiveModelBuilder, MisconceptionDetector, PersonalizedTeacher
print('modules imported')
```

```python
PA=PatternAnalyzer(); CMB=CognitiveModelBuilder(); MD=MisconceptionDetector(); PT=PersonalizedTeacher(); print('instances created')
```

```python
students=[{'student_id':'S001','responses':[{'rationale':'Voltage pushes current'},{'rationale':'Current flows'}]}]
features=PA.extract(students[0]['responses'])
print('features',features)
inf=CMB.build_offline(features)
print('inferred',inf)
det=MD.detect(inf,['voltage','current','resistance','ohm_law'])
print('detector',det)
rec={'target_nodes':['voltage','current'],'explanation_short':'Start with voltage'}
teach=PT.generate_offline(rec)
print('teaching',teach)
```
